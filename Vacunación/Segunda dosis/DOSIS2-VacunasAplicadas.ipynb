```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Obtiene las solamente las columnas orden_dosis y marca de la vacuna del dataset.
columnas = ['orden_dosis', 'vacuna']
try:
    data = pd.read_csv('../datos_nomivac_covid19.csv', usecols = columnas, skipinitialspace = True)
except Exception as Error:
    print('Hubo un error al leer el dataset', Error)
```

```python
# Obtiene todas las vacunas aplicadas de la segunda dosis.
primeraDosis = data.query('orden_dosis == 2')
```

```python

# Obtiene la cantidad de cada vacuna aplicada
sputnik = primeraDosis.query('vacuna == "Sputnik"').vacuna.count()
astrazeneca = primeraDosis.query('vacuna == "AstraZeneca"').vacuna.count()
sinopharm = primeraDosis.query('vacuna == "Sinopharm"').vacuna.count()
moderna = primeraDosis.query('vacuna == "Moderna"').vacuna.count()
```

```python
data = {'Sputnik': sputnik, 'Astrazeneca': astrazeneca, 'Sinopharm': sinopharm, 'Moderna': moderna}
vacunas = list(data.keys())
cantidad = list(data.values())

fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].bar(vacunas, cantidad)
axs[1].scatter(vacunas, cantidad)
fig.suptitle('Cantidad de vacunas aplicadas DOSIS 2')
```
